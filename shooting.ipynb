```python
import pandas as pd
import numpy as np
```

```python
#used http://wikitable2csv.ggor.de/ to convert wikitable to csv, very handy!
data = pd.read_csv("A:\dataset\shooting\shooting.csv")
data
```

```python
city_val = []
for city in data['Location'].str.split(","):
    city_val.append(city)
```

```python
#this is used to get counties of cities and state, I think there is a better API out there, this partially worked.
from geopy.geocoders import Nominatim
geolocator = Nominatim()
for location in city_val:
    loc = geolocator.geocode(location)
    try:
        city,county,state, zipcode, country = (loc.address.split(','))
    except:
        county = np.nan
    county_val.append(county)
data['County'] = pd.DataFrame(county_val)
# u'Chicago, Cook County, Illinois, United States of America'

# u'SF, California, United States of America'
```

```python
header = ['County','Location']
data.to_csv("A:\dataset\shooting\county.csv", columns = header)
```
