```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("A:/dataset/shooting/US mass shooting - Sheet1.csv")
data.head(5)
```

```python
data.columns
```

```python
data['Age '].describe()
```

```python
data['Age '].isnull().sum()/data['Age '].shape[0]*100
#29 % of age column with no values
```

```python
data['Age '].head(3)
```

```python
no_missing = data[data['Age '].notnull()]
no_missing['Age '].describe()
# The median age is 17, and the youngest is 6!
```

```python
no_missing['Age '].plot(kind = 'hist', bins = 10)
#Histogram of age of shooters
```

```python
no_missing['Age '].plot(kind = 'density')
```

```python
frequent_count = data['County'].value_counts()[data['County'].value_counts() > 1]
frequent_count.plot(kind = 'barh', figsize=(12, 12))
plt.xlabel('count of shooting')
plt.ylabel('County names')
```

```python
data.groupby('County').filter(lambda X : X['County'].value_counts() > 1)['Age '].median()
# Median age of shooters in counties where the shooting happened more than once
```
